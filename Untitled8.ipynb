```python
import pandas as pd

sensus = {
    'usia' : [63, 67, 67, 37, 41, 56, 62, 57, 63, 53],
    'sex' : [ 1, 1, 1, 1, 0, 1, 0, 0, 1, 1],
    'trestbps' : [145, 160, 120, 130, 130, 120, 140, 120, 130, 140],
     'chol' : [
        '233', '286', '229', '250', '204', '236', '268', '354', '254', '203'
        ]
}

sensus_df = pd.DataFrame(sensus)
sensus_df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for jk, d in sensus_df.groupby('chol'):
  ax.scatter(d['usia'], d['trestbps'], label=jk)

plt.legend(loc='upper left')
plt.title('Klasifikasi Penyakit Jantung')
plt.xlabel('usia')
plt.ylabel('trestbps')
plt.grid(True)
plt.show()
```

```python
import numpy as np

x_train = np.array(sensus_df[['usia', 'trestbps']])
y_train = np.array(sensus_df['sex'])

print(f'x_train:\n{x_train}\n')
print(f'y_train: {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n{y_train}')
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(x_train, y_train)
```

```python
age = 67
trestbps = 160
X_new = np.array([age, trestbps]).reshape(1, -1)
X_new
```

```python
y_new = model.predict(X_new)
y_new
```

```python
X_test = np.array([[160, 67], [145, 63], [150, 62], [140, 63]])
y_test = lb.transform(np.array(['pria', 'pria', 'wanita', 'wanita'])). flatten()

print(f'X_test:\n{X_test}\n')
print(f'y_test:\n{y_test}')
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)

print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import precision_score

prec = precision_score(y_test, y_pred)

print(f'Precision: {prec}')
```

```python
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred)

print(f'recall: {rec}')
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print(f'F1-score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)

print(f'Classification Report:\n {cls_report}')
```
